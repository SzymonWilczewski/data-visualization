```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv("share-of-individuals-using-the-internet.csv")

data_2017 = data.query("Year==2017")

px.choropleth(data_2017, locations="Code", color="% of population").show()
```

```python
data.sort_values(by=["Year"], inplace=True)

fig = px.choropleth(data, locations="Code", color="% of population", hover_name="Entity", hover_data={'Code': False, '% of population': True}, color_continuous_scale="Blues", animation_frame="Year")

fig.update_layout(title_text='Share of the population using the Internet, 1990-2017', title_font_family='Lato', geo=dict(showframe=False, projection_type='equirectangular'))

fig.show()
```
