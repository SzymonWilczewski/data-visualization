```python
from wordcloud import WordCloud, get_single_color_func
import matplotlib.pyplot as plt
from skimage.io import imread
import squarify
import matplotlib
from joypy import joyplot
import pandas as pd
```

```python
# Exercise 1.1

text=("Kamil,Tymoteusz,Filip,Dawid,Wojciech,Aleksandra,Robert,Paweł,Aleksandra,Piotr,Jakub,Michał,Mateusz,Kacper,Grzegorz,Paweł,Paweł,Maciej,Kamil,Piotr,Maciej,Maciej,Aleksander,Szymon,Filip,Tural")
wordcloud = WordCloud().generate(text)

plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```python
# Exercise 1.2

wordcloud = WordCloud(background_color='white').generate(text)
color_func1 = get_single_color_func('#4E79A7')

plt.figure(figsize=(10,10))
plt.imshow(wordcloud.recolor(color_func=color_func1, random_state=3))
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```python
# Exercise 1.3

mask = imread('circle.png')

wordcloud = WordCloud(mask=mask, background_color='white').generate(text)

plt.figure(figsize=(10,10))
plt.imshow(wordcloud.recolor(color_func=color_func1, random_state=3))
plt.axis("off")
plt.margins(x=0, y=0)
plt.savefig('WordCloud.png', dpi=300, transparent=True)
plt.show()
```

```python
# Exercise 2.1

sizes=[44, 19, 13, 12,9, 9, 3,3]
labels=["Australia",  "Norway",  "UK","unclaimed", "Chile","Argentina","New Zealand", "France"]

plt.figure(figsize=(10,5))
squarify.plot(sizes=sizes, label=labels, alpha=1)
plt.title("Who owns Antarctic?", size=20)
plt.show()
```

```python
# Exercise 2.2

matplotlib.rcParams['font.sans-serif'] = 'Lato'
matplotlib.rcParams['text.color'] = 'white'
matplotlib.rcParams['font.size'] = '14'
matplotlib.rcParams['font.weight'] = 'bold'

plt.figure(figsize=(10,5))
squarify.plot(sizes=sizes, label=labels, alpha=1, bar_kwargs=dict(linewidth=2, edgecolor="#FFFFFF"), color=['#0C3953',  '#094C72',  'tab:gray',  '#026AA7',  '#298FCA',  '#298FCA', '#8BBDD9', '#8BBDD9'])
plt.title("Who owns Antarctic?", size=20, color='black')
plt.axis('off')
plt.show()
```

```python
# Exercise 3.1.1

data = pd.read_csv('GdańskPogoda.csv', usecols=['DATA','min','max','YEAR'], sep=';')
ax, fig = joyplot(data, by='YEAR', column="max")
plt.title('GDAŃSK: average temperatures in March 2019-21', color='black')
plt.show()
```

```python
# Exercise 3.1.2

matplotlib.rcParams['font.sans-serif'] = 'Lato'
matplotlib.rcParams['font.size'] = '14'
matplotlib.rcParams['font.weight'] = 'bold'

data = pd.read_csv('GdańskPogoda.csv', usecols=['DATA','min','max','YEAR'], sep=';')
ax, fig = joyplot(data, by='YEAR', column="max", alpha=0.5, grid='x')
plt.title('GDAŃSK: average temperatures in March 2019-21', color='black')
plt.show()
```

```python
# Exercise 3.2

data = pd.read_csv('GdańskPogoda.csv', usecols=['DATA','min','max','YEAR'], sep=';')
ax, fig = joyplot(data, by='YEAR', column=["min", "max"], alpha=0.5, grid='x', color=['tab:blue', 'tab:red'])
plt.title('GDAŃSK: min and max temperatures in March 2019-21', color='black')
plt.show()
```
