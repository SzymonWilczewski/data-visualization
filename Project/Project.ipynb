```python
from skimage.io import imread
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'notebook_connected'
import plotly.graph_objects as go
import numpy as np
```

```python
# Chart 1

# Before

plt.figure(figsize=(14, 10))
plt.imshow(imread('Chart1.png')), plt.axis('off')
plt.show()

# After

def dynamic_title(i):
    return f'<b>Światowe składki ubezpieczeniowe w latach 2009-2018 (USD mld)<br>Razem: {dynamic_titles[i]}</b>'

chart1_df = pd.read_csv('Chart1.csv')
chart1_df.sort_values(by=['Year', 'Value'], inplace=True)

dynamic_titles = chart1_df.groupby(['Year'])['Value'].sum().reset_index(drop=True)

fig = px.bar(chart1_df, x='Value', y='Region', text='Value', animation_frame='Year', range_x=[0, chart1_df['Value'].max()], title=dynamic_title(0), width=800)
for i in range(len(fig.frames)):
    fig.frames[i]['layout'].update(title_text=dynamic_title(i))

fig.update_layout(font_family='Lato', hovermode=False, title_x=0.5, xaxis_title='', yaxis_title='')
fig['layout']['sliders'][0]['currentvalue']['visible'] = False
fig.show()
```

```python
# Chart 2

# Before

plt.figure(figsize=(14, 10))
plt.imshow(imread('Chart2.png')), plt.axis('off')
plt.show()

# After

chart2_df = pd.read_csv('Chart2.csv')

fig = go.Figure()

countries = chart2_df.Region.unique()
for country in countries:
    fig.add_trace(go.Pie(
        labels=chart2_df[chart2_df.Region == country].Branch,
        name=country,
        values=chart2_df[chart2_df.Region == country].Value,
        marker_colors=['darkblue', 'royalblue'],
        showlegend=False,
        text=['Dział I', 'Dział II']
    ))

updatemenu = []
buttons = []
for country in countries:
    buttons.append(dict(method='restyle',
                        label=country,
                        args=[{
                            'labels': [chart2_df[chart2_df.Region == country].Branch],
                            'values': [chart2_df[chart2_df.Region == country].Value]
                            }])
                  )

updatemenu=[]
your_menu=dict()
updatemenu.append(your_menu)
updatemenu[0]['buttons']=buttons
updatemenu[0]['active']=countries.size - 1

fig.update_layout(updatemenus=updatemenu, hovermode=False, title='Struktura składki przypisanej brutto krajów Unii Europejskiej w 2018 r.', width=800)
fig.show()
```

```python
# Chart 4

# Before

plt.figure(figsize=(14, 10))
plt.imshow(imread('Chart4.png')), plt.axis('off')
plt.show()

# After

chart1_df = pd.read_csv('Chart4.csv')
chart1_df.sort_values(by=['Year', 'Value'], inplace=True)

fig = px.bar(chart1_df, x='Value', y='Type', text='Value', animation_frame='Year', range_x=[0, 100], title='Zmiana struktury składki przypisanej brutto Działu I w Polsce w latach 2010-2019 w ujęciu realnym (%)', width=800)

fig.update_layout(font_family='Lato', hovermode=False, title_x=0.5, xaxis_title='', yaxis_title='')
fig['layout']['sliders'][0]['currentvalue']['visible'] = False
fig.show()
```
