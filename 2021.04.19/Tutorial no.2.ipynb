```python
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
import numpy as np
from matplotlib_venn import venn3, venn3_circles
import matplotlib
```

```python
# Exercise 1.1

# 1. create a table with data
GUI_1 = {'effectiveness': [8],'efficiency': [8],'user satisfaction': [7],'memorability': [9],
         'learnability': [9],'error handling': [8]}
table = pd.DataFrame(data=[GUI_1])

# 2. create a list of categories
categories=list(table)[0:]
N = len(categories)
 
# 3. create a list of values (values[:1] -> character from the beginning to position 1 (excluded))
values=table.loc[0].values.flatten().tolist()
values = values + values[:1]
 
# 4. create a list of angles
angles = [n*2*pi/N for n in range(N)]
angles += angles[:1]
   
# 5. plot the data using the polar coordinate system
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
# clockwise direction
# TODO: change the value of theta_offset to check what it does  
ax.set_theta_direction(-1)
ax.set_theta_offset(np.deg2rad(90))
ax.plot(angles, values, color='tab:blue', linewidth=2)
ax.fill(angles, values, 'tab:blue', alpha=0.2)
# TODO: change the value of rlabel_position to check what it does
ax.set_rlabel_position(90)
plt.xticks(angles[:-1], categories, color='#383428', size=15)
plt.yticks([2,4,6,8,10], ["2","4","6","8","10"], color='#606060', size=15, alpha=1, va = 'baseline')
plt.ylim(0, 10)
# nicely locate the category names 
for label, angle in zip(ax.get_xticklabels(), angles):
    if angle in (0, np.pi):
        label.set_horizontalalignment('center')
    elif 0 < angle < np.pi:
        label.set_horizontalalignment('left')
    else:
        label.set_horizontalalignment('right')
plt.show()
```

```python
# Exercise 1.2

matplotlib.rcParams['font.sans-serif'] = 'Lato'

GUI_1 = {'effectiveness': [8],'efficiency': [8],'user satisfaction': [7],'memorability': [9],
         'learnability': [9],'error handling': [8]}
GUI_2 = {'effectiveness': [5],'efficiency': [10],'user satisfaction': [6],'memorability': [5],
         'learnability': [10],'error handling': [9]}
table = pd.DataFrame(data=[GUI_1, GUI_2])

categories=list(table)[0:]
N = len(categories)

values_1=table.loc[0].values.flatten().tolist()
values_1 = values_1 + values_1[:1]

values_2=table.loc[1].values.flatten().tolist()
values_2 = values_2 + values_2[:1]

angles = [n*2*pi/N for n in range(N)]
angles += angles[:1]

plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
ax.set_theta_direction(-1)
ax.set_theta_offset(np.deg2rad(90))
ax.plot(angles, values_1, color='tab:orange', linewidth=2)
ax.fill(angles, values_1, 'tab:orange', alpha=0.2)
ax.plot(angles, values_2, color='tab:blue', linewidth=2)
ax.fill(angles, values_2, 'tab:blue', alpha=0.2)
ax.set_rlabel_position(90)
plt.xticks(angles[:-1], categories, color='#383428', size=15)
plt.yticks([2,4,6,8,10], ["2","4","6","8","10"], color='#606060', size=15, alpha=1, va = 'baseline')
plt.ylim(0, 10)
for label, angle in zip(ax.get_xticklabels(), angles):
    if angle in (0, np.pi):
        label.set_horizontalalignment('center')
    elif 0 < angle < np.pi:
        label.set_horizontalalignment('left')
    else:
        label.set_horizontalalignment('right')
start = 0.26
plt.figtext(start, 0.96, 'Usability Comparison: ', fontsize=20, fontweight='bold', color='black')
plt.figtext(start + 0.28, 0.96, 'WEB1', fontsize=20, fontweight='bold', color='tab:blue')
plt.figtext(start + 0.36, 0.96, ' vs. ', fontsize=20, fontweight='bold', color='black')
plt.figtext(start + 0.41, 0.96, 'WEB2', fontsize=20, fontweight='bold', color='tab:orange')
plt.show()
```

```python
# Exercise 2

matplotlib.rcParams['font.sans-serif'] = 'Lato'
matplotlib.rcParams['font.size'] = '20'
matplotlib.rcParams['text.color'] = 'white'

plt.figure(figsize=(15,15))

v=venn3(subsets = (10,10,2,10,2,2,2))

def setV(id, color, alpha, x, y, ha, text, text_color):
    v.get_patch_by_id(id).set_color(color)
    v.get_patch_by_id(id).set_alpha(alpha)
    v.get_label_by_id(id).set_x(x)
    v.get_label_by_id(id).set_y(y)
    v.get_label_by_id(id).set_ha(ha)
    v.get_label_by_id(id).set_text(text)
    v.get_label_by_id(id).set_color(text_color)

setV('100','#EE2124', 1, -0.59, 0.15, 'left', '\n \n proper collection & \n organization of data \n with a purpose ', 'white')
setV('010','#6BBE45', 1, 0.59, 0.15, 'right', '\n \n accurate, simple \n & attractive \n representation of data ', 'white')
setV('001','#3955A7', 1, 0, -0.5, 'center', ' impactful & interesting \n visuals with balanced \n form & function ', 'white')
setV('110','#F7ED14', 1, 0, 0.2, 'center', ' Rigid & \n Complex ', 'black')
setV('011','#70CDDE', 1, 0.19, -0.11, 'center', ' Lack of \n Complexity ', 'black')
setV('101','#C97BB5', 1, -0.19, -0.11, 'center', ' Beautiful \n but Illegible ', 'black')
setV('111','#FFFFFF', 1, 0, 0, 'center', ' A Good Data \n Visualization ', 'black')

plt.title('What Makes a Good\nDATA VISUALIZATION', size=36, color='black', fontweight='bold')

plt.text(-0.57, 0.2, "DATA", weight='bold', color="#FFFFFF", size=35)
plt.text(0.27, 0.2, "FUNCTION", weight='bold', color="#FFFFFF", size=35)
plt.text(-0.1, -0.4, "DESIGN", weight='bold', color="#FFFFFF", size=35)

plt.show()
```

```python
# Exercise 3.2

matplotlib.rcParams['axes.edgecolor'] = "#FFFFFF"
matplotlib.rcParams['xtick.color'] = "#FFFFFF"
matplotlib.rcParams['ytick.color'] = "#FFFFFF"
matplotlib.rcParams['figure.facecolor'] = "#FFFFFF"
matplotlib.rcParams['text.color'] = "#808080"
matplotlib.rcParams['font.sans-serif'] = 'Lato'

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(24,12))

def waffle_chart(country, percentage, value, ax):
    data = np.fliplr(np.append(np.ones((100-percentage)), np.full((percentage), value)).reshape((10, 10)))
    axes[ax].set_title(country, size=30, color='#383428', weight="bold")
    axes[ax].matshow(data, cmap='tab20c', vmin=0, vmax=1)
    axes[ax].set_xticks(np.arange(-.5, (10), 1), minor=True)
    axes[ax].set_yticks(np.arange(-.5, (10), 1), minor=True)
    axes[ax].grid(which='minor', color='w', linestyle='-', linewidth=2)
    axes[ax].set_xticks([])
    axes[ax].set_yticks([])
    axes[ax].set_xlabel(str(percentage) + '%', color=plt.cm.tab20c(value), fontweight='bold', size=32, labelpad=-172)

waffle_chart('Germany', 21, 0, (0, 0))
waffle_chart('Lithuania', 20, 0, (0, 1))
waffle_chart('Czech Republic', 20, 0, (0, 2))
waffle_chart('Poland', 18, 0.2, (0, 3))
waffle_chart('Ukraine', 17, 0, (1, 0))
waffle_chart('Slovakia', 16, 0, (1, 1))
waffle_chart('Belarus', 15, 0, (1, 2))
waffle_chart('Russia', 15, 0, (1, 3))

start = 0.17
plt.figtext(start, 0.96, 'The Percentage of Older Adults (65+) in ', fontsize=40, color='black')
plt.figtext(start + 0.41, 0.96, 'Poland', fontsize=40, color=plt.cm.tab20c(0.2))
plt.figtext(start + 0.48, 0.96, ' and it\'s ', fontsize=40, color='black')
plt.figtext(start + 0.565, 0.96, 'Neighbours', fontsize=40, color=plt.cm.tab20c(0))

plt.show()
```
